Причина наличия стоимости продукции в таблице данных выше 500$ имеет как негативный, так и позитивный сценарий.

Позитивный сценарий: это обосновонная стоимость особого сорта вина.

Поэтому, есть высокая вероятность, что при соблюдении определенных условий, будет рассматриваться возможность удаления данных единиц продукции дороже 300 и 200$.

Оставим в исследовании только Красные и Белые вина.

```python
data_regr=data4[['color','points','price']][data4['color']!=0].dropna().reset_index()
print('Проверка на уникальные значения в столбце \'color\'')
print(data_regr.color.unique())
print('Информация о данных в DataFrame.data_regr:')
print(data_regr.info())
print('Информация о данных в DataFrame.data_regr после приведения числовых значений к типу float64:')
data_regr=data_regr.astype({'points':'float64'})
print(data_regr.info())
```

```python
#загружаем данные для регрессии
X,y=data_regr[['points']],data_regr.price
#создаем данные для обучения и тестирования
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

Далее на обучающей выборке строим регрессионную модель, зависимость между рейтингом'points'(зависимая переменная) и стоимостью'price'. 

```python
model=BayesianRidge()
viewer=ResidualsPlot(model)
viewer.fit(X_train, y_train)
viewer.score(X_test,y_test)
plt.figtext(0, -0.1, "рис.22 'Регрессионная модель зависимости рейтинга от стоимости(обучающая и тестовая выборка)", fontsize =10)
viewer.show()
```
